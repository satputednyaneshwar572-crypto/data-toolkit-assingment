```python
#Q2 write a pandas script to find the mean of a specific column in a dataframe 
import pandas as pd
df = pd.DataFrame({"a":[1, 2, 3], 'b':[5, 6, 7]})
me = df['a'].mean()
print(me)
```

```python
#creat scatter plot using matplotly
import matplotlib.pyplot as plt
x = [1, 7, 3]
y = [5, 6, 7]
plt. scatter(x, y, color = "red", marker = 'o')
plt.xlabel("x values")
plt.ylabel('y values')
plt.title("scatter plot")
plt.show()
```

```python
#Q4 how do you calculate the correlation matrix using seaborn and visualize it with a heatmap
import seaborn as sns
df = pd.DataFrame({"a":[1, 2, 4,], 'b':[5, 6, 7]})
ma = df.corr()
print(ma)
plt.figure(figsize=(6, 4))
sns.heatmap(ma, linewidth=0.6)
plt.title("heat map")
plt.show()
```

```python
#Q5 generat a bar plot using matplotlib
import plotly.express as px
df = pd.DataFrame({'ctg':['a', 'b', 'c'], 'val':[3, 4, 5]})
fig = px.bar(df, x="ctg", y = "val")
fig.show()
```

```python
#Q7 write a program to perform element wise multiplication of two numpy arrays
a = np.array([1, 2, 3, 4])
b = np.array([6, 7, 8, 9])
c = a*b
print(c)
```

```python
#Q8 creat a line plot with multiple lines using matplotlib
x = np.linspace(4, 5, 34)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure(figsize= (7, 3))
plt.plot(x, y1, label = "sinx", linewidth = 3)
plt.plot(x, y2, label= "cosx", linewidth = 3)
plt.xlabel("x axis")
plt.ylabel("y label")
plt.title("multiple line")
plt.grid(True)
plt.legend()
plt.show()
```
